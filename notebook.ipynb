## Visualization & Explanatory Analysis

### Pertanyaan 1:

```python
plt.figure(figsize=(14, 7))
sns.lineplot(data=hour_data, x='hr', y='cnt', hue='weathersit')
plt.title('Penyewaan Sepeda berdasarkan Jam dan Kondisi Cuaca')
plt.xlabel('Jam')
plt.ylabel('Jumlah Penyewaan')
plt.show()
sns.histplot(hour_data['cnt'])
plt.show()
```

- Tujuan: Membuat grafik garis (line plot) untuk menunjukkan bagaimana jumlah penyewaan sepeda (cnt) bervariasi berdasarkan jam (hr) dan kondisi cuaca (weathersit).
- Kegunaan: Memahami pola penggunaan sepeda selama satu hari dan bagaimana cuaca mempengaruhi jumlah penyewaan.
